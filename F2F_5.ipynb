````python
def mean(data):
  n = len(data)
  mean = sum(data) / n
  return mean

def variance(data):
   
  n = len(data)
   
  mean = sum(data) / n
   
  deviations = [(x - mean) ** 2 for x in data]
    
  variance = sum(deviations) / n
  return variance

def stdev(data):
  import math
  var = variance(data)
  std_dev = math.sqrt(var)
  return std_dev
````

````python
data = {'Mean': [mean(df.AGE),mean(df.GP), mean(df.MPG)], 'Std. Deviation': [stdev(df.AGE), mean(df.GP), mean(df.MPG)], 'Variance': [variance(df.AGE), variance(df.GP), variance(df.MPG)]}
````
