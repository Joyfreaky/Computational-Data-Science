```python
def pluralize(word):
    '''This function takes in an argument as string, and check the if the string is an empty string, Plural or singular, and proper noun.  
    if the string is singular, it pluralizes the string and returns the status of conversion as sucess, failure or proper noun and the plural word.
    The function doesnot pluralize the proper noun in this case. '''
    words_store = []                             # Initialisation of variables
    status = ''
    word_in_plural = ''

    with open("proper_nouns.txt", "r") as file:   # Read the file Proper_noun.txt
        words_store = file.readlines()
        file.close() 

    for i in range(len(words_store)):            # Storing the words in the list. 
        words_store[i] = words_store[i][:-1]

    if word == '':                               # Conditional statement to check, if the words are empty string, plural, singular or proper noun
        status = 'empty_string'
        word_in_plural = 'plural'
        return  {word_in_plural: '', "status": status}  # Check Empty String and return the status and string
    elif word[-1] == 's':
        word_in_plural = word
        status = 'already_in_plural'
        return {'plural': word, "status": status}       # Check if the string is ending with s and return the status and string.
    elif word.lower() in words_store :
        word_in_plural = word
        status =  'proper_noun'                         # Check if the string is proper noun or not.
       

    else:                                             # If none of the above then convert the string into its plural form.
        if word[-1] == 'a' or word[-1] == 'e' or word[-1] == 'i' or word[-1] == 'o' or word[-1] == 'u':
            word_in_plural = word + 's'
        elif  word[-1] == 'y' and word[-2] != 'a' and word[-2] != 'o' and word[-2] != 'i' and word[-2] != 'e' and word[-2] != 'u':
            word_in_plural = word[:-1] + 'ies'
        elif word[-1] == 'f':
            word_in_plural = word[:-1] + 'ves'
        elif word[-2:] == 'sh' or word[-2:] == 'ch'  or word[-2:] == 'zz':
            word_in_plural = word + 'es'
        else:
            word_in_plural  = word + 's' 
            status = "success"
  
    return {"plural": word_in_plural, "status": status}  

        



### --- IMPORTANT: DO NOT REMOVE OR CHANGE THE CODE BELOW ---
TEST_CASES = """failure
food
Zulma
injury
elf
buzz
computers
PCs

highway
presentation
pouch
adam""".split('\n')

if __name__ == '__main__':

  for test_noun in TEST_CASES:

    print(test_noun,'-->',pluralize(test_noun))
    print('----')

### --- The outputs of test cases are given below. Please compare with your results.
# ************************************************************************************
# failure --> {'plural': 'failures', 'status': 'success'}
# ----
# food --> {'plural': 'foods', 'status': 'success'}
# ----
# Zulma --> {'plural': 'Zulma', 'status': 'proper_noun'}
# ----
# injury --> {'plural': 'injuries', 'status': 'success'}
# ----
# elf --> {'plural': 'elves', 'status': 'success'}
# ----
# buzz --> {'plural': 'buzzes', 'status': 'success'}
# ----
# computers --> {'plural': 'computers', 'status': 'already_in_plural'}
# ----
# PCs --> {'plural': 'PCs', 'status': 'already_in_plural'}
# ----
#  --> {'plural': '', 'status': 'empty_string'}
# ----
# highway --> {'plural': 'highways', 'status': 'success'}
# ----
# presentation --> {'plural': 'presentations', 'status': 'success'}
# ----
# pouch --> {'plural': 'pouches', 'status': 'success'}
# ----
# adam --> {'plural': 'adam', 'status': 'proper_noun'}
# ----
```
